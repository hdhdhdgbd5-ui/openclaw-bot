```python
# Cell 3: Start Ollama server in background
import subprocess
import time
import os
import threading

# Set environment variables
os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
os.environ['OLLAMA_ORIGINS'] = '*'

def start_ollama():
    """Start Ollama server"""
    process = subprocess.Popen(
        ['ollama', 'serve'],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE
    )
    return process

# Start server
print('🚀 Starting Ollama server...')
ollama_process = start_ollama()
time.sleep(15)  # Wait for server initialization
print('✅ Ollama server started!')
print(f'Process ID: {ollama_process.pid}')
```

```python
# Cell 7: Setup ngrok tunnel (FREE - no auth token needed for basic use)
from pyngrok import ngrok
import time

# Kill any existing tunnels
ngrok.kill()

# Create tunnel to Ollama server
public_url = ngrok.connect(11434, "http")
print(f"🌐 Ollama public URL: {public_url}")
print(f"")
print(f"⚠️  IMPORTANT: Copy this URL and give it to OpenClaw!")
print(f"The URL will change each time you restart the notebook.")
```

```python
# Cell 8: Keep-alive script (run this to keep session active)
import time
from datetime import datetime

print("💓 Keep-alive started. Press Stop to end.")
print("This prevents Kaggle from timing out.")

while True:
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{timestamp}] Still running... Ollama is alive!")
    time.sleep(300)  # Print every 5 minutes
```
